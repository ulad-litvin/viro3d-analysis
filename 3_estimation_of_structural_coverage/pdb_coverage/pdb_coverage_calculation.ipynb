# Summary PDB template coverage info

- Use JSON files: ef_protein_plddt_pdb_template_coverage.json and cf_protein_plddt_pdb_template_coverage.json

```python
import statistics

def plddt_mean(plddt: list, round_to: int = 1):
    '''
    Calculate the average pLDDT score.
    '''
    return round(statistics.mean(plddt), round_to)

def plddt_coverage_aa_number(plddt: list, lower_threshold: float,
                             upper_threshold: float, include_upper_threshold: bool = False):
    '''
    Calculate the coverage of a protein sequence in number of AA
    based on a pLDDT thresholds and return the number of AA.
    '''
    if include_upper_threshold:
        return len([i for i in plddt if i >= lower_threshold and i <= upper_threshold])
    else:
        return len([i for i in plddt if i >= lower_threshold and i < upper_threshold])

def plddt_coverage_across_default_categories(plddt: list):
    '''
    Calculate the coverage of a protein sequence in number of AA
    based on default pLDDT categories: [0, 50), [50, 70), [70, 90), [90, 100] -
    and return a list with the number of AA for each category.
    '''
    coverage = []
    coverage.append(plddt_coverage_aa_number(plddt, 0, 50))
    coverage.append(plddt_coverage_aa_number(plddt, 50, 70))
    coverage.append(plddt_coverage_aa_number(plddt, 70, 90))
    coverage.append(plddt_coverage_aa_number(plddt, 90, 100, include_upper_threshold=True))
    return coverage
```

# Calculate PDB coverage and save it in json

- Use JSON files with protein info:
    - ./cf_protein_plddt_coverage.json
    - ./ef_protein_plddt_coverage.json
- Use JSON files with training data cut-off info:
    - ./pdb_ids_released_before_2018-04-30.json
    - ./pdb_ids_released_before_2020-05-01.json
- Use TSV files with Mmseqs2 easy-search results against PDB 2024-02-20 release:
    - ../mmseqs2_analysis_pdb/all_proteins_pdb_results.tsv
- Produce JSON files with PDB structure info (95% seq identity to PDB 2024-02-20) and tempalte info (30% seq identity to PDB 2018-04-30 and 2020-05-01 for ColabFold and ESMFold respectively):
    - ./cf_protein_plddt_pdb_template_coverage.json
    - ./ef_protein_plddt_pdb_template_coverage.json

### JSON file structure:

- 71,274 items (matches the number of protein records)
- key: protein record_id (eg. 'ASO76148.1_10247')
    - value_1 key: 'record_id'
    - value_1 value: list of record_id (eg. 'ASO76148.1_10247') that cover this protein or its region structurally
    - value_2 key: 'plddt'
    - value_2 value: list of per residue pLDDT scores (eg. 76.0)
    - value_3 key: 'pdb_id_95%_2024-02-20'
    - value_3 value: list of PDB IDs (eg. mom3) from PDB 2024-02-20 that cover this protein or its region structurally with 95% sequence identity
    - value_4 key: 'pdb_id_30%_2020-05-01' (ESMFold only)
    - value_4 value: list of PDB IDs (eg. mom3) from PDB 2020-05-01 (ESMFold training data cut-off) that cover this protein or its region structurally with 30% sequence identity (template)
    - value_5 key: 'pdb_id_30%_2018-04-30' (ColabFold only)
    - value_4 value: list of PDB IDs (eg. mom3) from PDB 2018-04-30 (ColabFold training data cut-off) that cover this protein or its region structurally with 30% sequence identity (template)

```python
'''
ESMFold coverage by PDB entries with 30% sequence identity
from the PDB release 2020-05-01
'''

n = 0

for key, value in poly_json.items():

    n += 1
    header = key
    protlen = df_coverage.loc[df_coverage['header'] == header, 'protlen'].values[0]
    pdb_id_list = [np.nan] * protlen

    if header in df_pdb['query'].values:

        df_pdb_sub = df_pdb[df_pdb['query'] == header].sort_values(by=['alnlen'], ascending=False)

        for index, row in df_pdb_sub.iterrows():

            pdb_id = row['target_short']
            start = row['qstart']
            end = row['qend']
            
            for i in range(start, end+1):
                if pdb_id_list[i-1] is np.nan:
                    try:
                        pdb_id_list[i-1] = pdb_id
                    except IndexError:
                        print('ERROR: ', header, end - start + 1, len(pdb_id_list))
                    
    value['pdb_id_30%_2020-05-01'] = pdb_id_list
    pdb_coverage = len(pdb_id_list) - pdb_id_list.count(np.nan)
    print(n, header, round(pdb_coverage/ protlen, 2), pdb_coverage)
```
