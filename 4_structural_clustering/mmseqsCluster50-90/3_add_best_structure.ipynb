```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_meta = "./mmseqsCluster50-90_meta.csv"
file_stats = "./mmseqsCluster50-90_stats_with_best_structure.csv"

df_meta = pd.read_csv(file_meta)
df_stats = pd.read_csv(file_stats)
```

```python
# Check that the structure representative is indeed the member of the cluster

n = 0

for index, row in df_stats.iterrows():

    cluster_rep = row["cluster_representative"]
    cluster_structure_rep = row["structure_representative_record_id"]

    if cluster_structure_rep not in df_meta[df_meta["cluster_representative"] == cluster_rep]['cluster_member'].values:
        n += 1
        print(n, cluster_rep, cluster_structure_rep)
```

```python
for index, row in df_stats.iterrows():
    
    cluster_id = row["cluster_id"]
    cluster_rep = row["cluster_representative"]
    cluster_size = row["cluster_size"]
    cluster_pfam_desc = row["top_pfam_desc"]
    cluster_pfam_ratio = row["top_pfam_ratio"]
    
    df_sub = df_meta[df_meta["cluster_representative"] == cluster_rep].sort_values(by=["plddt", "ptm"], ascending=False)
    
    best_member_record_id = df_sub.iloc[0]["cluster_member"]
    best_member_class = df_sub.iloc[0]["member_class"]
    best_member_plddt = df_sub.iloc[0]["plddt"]
    best_member_pfam_desc = df_sub.iloc[0]["pfam_desc"]
    best_member_genbank_name = df_sub.iloc[0]["genbank_name"]
    
    df_stats.loc[index, "structure_representative_class"] = best_member_class
    df_stats.loc[index, "structure_representative_record_id"] = best_member_record_id
    df_stats.loc[index, "structure_representative_plddt"] = best_member_plddt
    df_stats.loc[index, "structure_representative_pfam_desc"] = best_member_pfam_desc
    df_stats.loc[index, "structure_representative_genbank_name"] = best_member_genbank_name
    
    print(cluster_id, best_member_record_id, best_member_pfam_desc, best_member_plddt)
```
