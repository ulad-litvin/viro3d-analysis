```python
#only keep viruses with genbank accession
msl38_v2_vi_gb = msl38_v2_vi.dropna(subset=['Virus GENBANK accession'])
msl38_v2_vi_gb
```

```python
#make df of all genbank accessions, accounting for segments
segmentacc = {'acc':[], 'seg':[], 'Sort':[]}
for a in list(msl38_v2_vi_gb['Virus GENBANK accession']):
    sta = a
    a = a.replace(' ', '')
    if (';' not in a) and (':' not in a):
        segmentacc['acc'].append(a)
        segmentacc['seg'].append('')
        segmentacc['Sort'].append(list(msl38_v2_vi_gb[msl38_v2_vi_gb['Virus GENBANK accession'] == sta].Sort)[0])
    else:
        segc = 0
        for aseg in a.split(';'):
#             print(a)
#             print(aseg)
            if ':' in aseg:
                segmentacc['acc'].append(aseg.split(':')[1])
                segmentacc['seg'].append(aseg.split(':')[0])
                segmentacc['Sort'].append(list(msl38_v2_vi_gb[msl38_v2_vi_gb['Virus GENBANK accession'] == sta].Sort)[0])            
            else:
                segmentacc['acc'].append(aseg)
                #if there's no segment annotation just enumerate
                segmentacc['seg'].append('Seg' + str(segc))
                segc = segc+1
                segmentacc['Sort'].append(list(msl38_v2_vi_gb[msl38_v2_vi_gb['Virus GENBANK accession'] == sta].Sort)[0])
                
msl38_v2_vi_gb_ntacc = pd.DataFrame(segmentacc)

#manual edit
msl38_v2_vi_gb_ntacc['acc'] = msl38_v2_vi_gb_ntacc['acc'].str.replace("\(2253.10260\)", '')

# msl38_v2_vi_gb_ntacc.to_csv('./msl38v2/nt_gb_msl38v2.csv', index =False)
msl38_v2_vi_gb_ntacc
```

```python
msl38_v2_vi_gb_ntdic = SeqIO.to_dict(SeqIO.parse('nt_gb_msl38v2.gbk', 'gb'))

#ensure that all accessions have made it to the gbk file
print(len(msl38_v2_vi_gb_ntdic))
print(set([x.split('.')[0] for x in list(msl38_v2_vi_gb_ntdic)]) - set(msl38_v2_vi_gb_ntacc.acc))
print(set(msl38_v2_vi_gb_ntacc.acc)- set([x.split('.')[0] for x in list(msl38_v2_vi_gb_ntdic)]) )
```

```python
infoquals = {'gene', 'note', 'product', 'protein_id', 'translation'}
msl38_v2_vi_gb_cdsacc = {x:[] for x in infoquals}
msl38_v2_vi_gb_cdsacc.update({'host':[], 'nt_acc':[], 'acc':[], 'coordinates':[]})


for r in list(msl38_v2_vi_gb_ntdic):
    thehost = ''
    for f in msl38_v2_vi_gb_ntdic[r].features:
        if f.type == 'source':
#             print(list(f.qualifiers))
            if 'host' in list(f.qualifiers):
                thehost = f.qualifiers['host'][0]
        elif f.type == 'CDS':
#             print(infoquals.intersection(set(f.qualifiers)))
            for q in infoquals.intersection(set(f.qualifiers)):
                #checked that they all have length of 1
                msl38_v2_vi_gb_cdsacc[q].append(f.qualifiers[q][0])
            #add gaps to columns that are not in the entry
            for q in infoquals - infoquals.intersection(set(f.qualifiers)):
                msl38_v2_vi_gb_cdsacc[q].append('')
            msl38_v2_vi_gb_cdsacc['nt_acc'].append(r)
            msl38_v2_vi_gb_cdsacc['acc'].append(r.split('.')[0])
            msl38_v2_vi_gb_cdsacc['host'].append(thehost)
            msl38_v2_vi_gb_cdsacc['coordinates'].append(str(f.location))
            
msl38_v2_vi_gb_cds_df = pd.DataFrame(msl38_v2_vi_gb_cdsacc)

msl38_v2_vi_gb_cds_df = msl38_v2_vi_gb_cds_df[msl38_v2_vi_gb_cds_df.protein_id != '']
msl38_v2_vi_gb_cds_df['protlen'] = [len(x) for x in list(msl38_v2_vi_gb_cds_df.translation)]


msl38_v2_vi_gb_cds_df.to_csv('./msl38v2/nt_gb_msl38v2_cds.csv', index=False)

msl38_v2_vi_gb_cds_df

# #.tail(50)
```

```python
mat_pept_dic = {'protein_id':[], 'mat_pept_seq':[], 'note':[], 'mat_pept_id':[], 'protlen':[], 'coordinates':[]}
for r in list(msl38_v2_vi_gb_aadic):
#     print(msl38_v2_vi_gb_aadic[r])
    c = 0
    for f in msl38_v2_vi_gb_aadic[r].features:
        if f.type == 'mat_peptide':
#             print(str(msl38_v2_vi_gb_aadic[r].seq))
#             print(f.location)
#             print(str(msl38_v2_vi_gb_aadic[r].seq)[f.location.start:f.location.end])
            mat_pept_dic['protein_id'].append(r)
            mat_pept_dic['mat_pept_seq'].append(str(msl38_v2_vi_gb_aadic[r].seq)[f.location.start:f.location.end])
            mat_pept_dic['protlen'].append(len(str(msl38_v2_vi_gb_aadic[r].seq)[f.location.start:f.location.end]))
            mat_pept_dic['note'].append(f.qualifiers['product'][0])
            mat_pept_dic['mat_pept_id'].append(r + '.'+ str(c+1))
            mat_pept_dic['coordinates'].append(str(f.location))
            c =c +1
            

mat_pept_df = pd.DataFrame(mat_pept_dic)
mat_pept_df
```

```python
#add regions for everything >= 2000aa
reg_dic = {'protein_id':[], 'reg_seq':[], 'note':[], 'reg_id':[], 'protlen':[], 'coordinates':[]}
for r in list(msl38_v2_vi_gb_aadic):
#     print(msl38_v2_vi_gb_aadic[r])
    c = 0
    if len(msl38_v2_vi_gb_aadic[r].seq) >= 2000:
        for f in msl38_v2_vi_gb_aadic[r].features:
            if f.type == 'Region':
#                 print(f)
    #             print(str(msl38_v2_vi_gb_aadic[r].seq))
    #             print(f.location)
    #             print(str(msl38_v2_vi_gb_aadic[r].seq)[f.location.start:f.location.end])
                reg_dic['protein_id'].append(r)
                reg_dic['reg_seq'].append(str(msl38_v2_vi_gb_aadic[r].seq)[f.location.start:f.location.end])
                reg_dic['protlen'].append(len(str(msl38_v2_vi_gb_aadic[r].seq)[f.location.start:f.location.end]))
                reg_dic['coordinates'].append(str(f.location))
                reg_desc = ''
                for did in ['db_xref', 'region_name', 'note']:
                    if did in list(f.qualifiers):
                        reg_desc = reg_desc + f.qualifiers[did][0] + '|'
                reg_dic['note'].append(reg_desc[:-1])
#                 reg_dic['note'].append(f.qualifiers['region_name'][0])
                reg_dic['reg_id'].append(r + '.1.'+ str(c+1))
                c =c +1
            

reg_df = pd.DataFrame(reg_dic)
reg_df
```

## ESMfold tests

```python
virofold_full = pd.read_csv('virosphere-fold-v1_fulldata.csv')
```

```python
virofold100_1000 = virofold_full[(100 <= virofold_full.protlen) &  (virofold_full.protlen <= 1000)]
```

```python
#take 10 samples of 100 proteins each with lengths between 100 and 1000 amino acids to speed test ESMfold on alpha2
for i in range(10):
    fasta = ''
    ransam = random.sample(list(virofold100_1000.header), 100)
    for x in ransam:
        fasta = fasta + '>%s\n%s\n'%(x, list(virofold100_1000[virofold100_1000.header == x].translation)[0])
    with open('esmfold_timetest/esmfold_test%i.fas'%i, 'w') as out:
        out.write(fasta[:-1])
```
